### 2. Examine the State and Action Spaces

### 3. Take Random Actions in the Environment

In the next code cell, you will learn how to use the Python API to control the agent and receive feedback from the environment.

Once this cell is executed, you will watch the agent's performance, if it selects an action (uniformly) at random with each time step.  A window should pop up that allows you to observe the agent, as it moves through the environment.  

Of course, you'll have to change the code so that the agent is able to use its experience to gradually choose better actions when interacting with the environment!

````python
env_info = env.reset(train_mode=False)[brain_name] # reset the environment
state = env_info.visual_observations[0]            # get the current state
score = 0                                          # initialize the score
while True:
    action = np.random.randint(action_size)        # select an action
    env_info = env.step(action)[brain_name]        # send the action to the environment
    next_state = env_info.visual_observations[0]   # get the next state
    reward = env_info.rewards[0]                   # get the reward
    done = env_info.local_done[0]                  # see if episode has finished
    score += reward                                # update the score
    state = next_state                             # roll over the state to next time step
    if done:                                       # exit loop if episode finished
        break
    
print("Score: {}".format(score))
````

### 4. Train an agent to solve the environment

Now it's your turn to train your own agent to solve the environment!  When training the environment, set `train_mode=True`, so that the line for resetting the environment looks like the following:
```python
env_info = env.reset(train_mode=True)[brain_name]
```

````python
from unityagents import UnityEnvironment
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from dqn_agent import Agent
from collections import deque
import torch
````

````python
# Load environment
env = UnityEnvironment(file_name="./VisualBanana_Windows_x86_64/Banana.exe")
````

````python
# get the default brain
brain_name = env.brain_names[0]
brain = env.brains[brain_name]

# Get action state and action size
env_info = env.reset(train_mode=True)[brain_name]
state_size = np.squeeze(env_info.visual_observations[0])
state_size = np.transpose(state_size, (2, 0, 1)).shape
action_size = brain.vector_action_space_size

# Init agent
agent = Agent(state_size=state_size, action_size=action_size, seed=42, convolutional=True)
````

````python
# Set learning params
n_episodes = 2000
max_t = 1000
eps_start = 1.0
eps_end = 0.01
eps_decay = 0.995

# Init learning
scores = []                        # list containing scores from each episode
scores_window = deque(maxlen=100)  # last 100 scores
eps = eps_start                    # initialize epsilon

# Loop of episodes
for i_episode in range(1, n_episodes+1):
    # Reset environment
    env_info = env.reset(train_mode=True)[brain_name]
    # Get init state and score
    state = np.squeeze(env_info.visual_observations[0])
    state = np.transpose(state, (2, 0, 1))
    score = 0

    # Episode loop
    for t in range(max_t):
        # Get action to perform given the state
        action = agent.act(state, eps)
        action = action.astype(np.int32)

        # Get new state and return
        env_info = env.step(action)[brain_name]
        next_state = np.squeeze(env_info.visual_observations[0])
        next_state = np.transpose(next_state, (2, 0, 1))
        reward = env_info.rewards[0]
        done = env_info.local_done[0]

        # Agent learns
        agent.step(np.expand_dims(state, 0), action, reward, np.expand_dims(next_state, 0), done)

        # Prepare state and score for next iteration
        state = next_state
        score += reward
        if done:
            break

    # Append score to scores lists
    scores_window.append(score)
    scores.append(score)
    # Decrease epsilon of eps-greedy policy
    eps = max(eps_end, eps_decay*eps)

    # Plot results of episodes and save if score surpass our target
    print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)), end="")
    if i_episode % 100 == 0:
        print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
    if np.mean(scores_window)>=10.0:
        print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_window)))
        torch.save(agent.qnetwork_local.state_dict(), 'visual_qnet.pth')
        break
````

````python
# Store the model anyway
torch.save(agent.qnetwork_local.state_dict(), 'visual_qnet.pth')
````

````python
import matplotlib.pyplot as plt

# plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
````

### 5. Watch the agent perform

````python
env_info = env.reset(train_mode=False)[brain_name] # reset the environment
state = np.squeeze(env_info.visual_observations[0])# get the current state
state = np.transpose(state, (2, 0, 1))
score = 0                                          # initialize the score
while True:
    # Get action to perform given the state
    action = agent.act(state)
    action = action.astype(np.int32)

    # Get new state and return
    env_info = env.step(action)[brain_name]
    next_state = np.squeeze(env_info.visual_observations[0])
    next_state = np.transpose(next_state, (2, 0, 1))
    reward = env_info.rewards[0]
    done = env_info.local_done[0]

    # Prepare state and score for next iteration
    state = next_state
    score += reward
    if done:
        break

print("Score: {}".format(score))
````
